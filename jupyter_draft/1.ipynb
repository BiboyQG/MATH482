```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)

# Define the constraints
constraint1 = 6 * X + 5 * Y <= 19
constraint2 = Y <= 4 * X + 9
constraint3 = 2 * X - 7 * Y <= 15

# Find the feasible region (intersection of all constraints)
feasible_region = constraint1 & constraint2 & constraint3

# Create the plot
plt.figure(figsize=(10, 8))

# Plot the feasible region
plt.imshow(
    feasible_region,
    extent=[x.min(), x.max(), y.min(), y.max()],
    origin="lower",
    cmap="Blues",
    alpha=0.3,
)

# Plot the constraint lines
x_vals = np.array([x.min(), x.max()])

# 6x + 5y = 19
plt.plot(x_vals, (19 - 6 * x_vals) / 5, "r-", label="6x + 5y = 19")

# y = 4x + 9
plt.plot(x_vals, 4 * x_vals + 9, "g-", label="y = 4x + 9")

# 2x - 7y = 15
plt.plot(x_vals, (2 * x_vals - 15) / 7, "b-", label="2x - 7y = 15")

# Customize the plot
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Feasible Region for Linear Program")
plt.legend()
plt.axis("equal")

# Show the plot
plt.show()
```
